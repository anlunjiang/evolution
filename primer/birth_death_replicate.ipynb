# Why do things exist
> Setting the stage for evolution

```python
import random 
import matplotlib.pyplot as plt
```

```python
# With each frame (iteration) one blob will be created and each living 
# blob will have a chance of dying 

class create_blob(object):
    spawn_count = 0
    def __init__(self, nm):
        self.id = nm
        create_blob.spawn_count += 1
    def death(self):
        create_blob.spawn_count -= 1

def cycle_blobs(birth_chance, death_chance, replicate_chance, iterations):
    
    create_blob.spawn_count = 0
    blob_list = []
    spawncount_list = []
    
    for i in range(0, iterations):
        if random.random() <= birth_chance:
            blob = create_blob(i)
            blob_list.append(blob)
        
        for j in blob_list:
            if random.random() <= replicate_chance:
                blob = create_blob(len(blob_list) + 1)
                blob_list.append(blob)
            if random.random() <= death_chance:
                j.death()
                blob_list.remove(j)
        spawncount_list.append(create_blob.spawn_count)

    plt.plot(list(range(1, iterations + 1)), spawncount_list)
    plt.xlabel('Iteration')
    plt.ylabel('Number')
    plt.title('Animation of Simulation')
    avg = sum(spawncount_list) / len(spawncount_list)
    print(f'Birth chance: {birth_chance}')
    print(f'Replication chance: {replicate_chance}')
    print(f'Death chance: {death_chance}')
    print(f'Average number of blobs: {avg}')
```

```python
# Expected number of blobs to exist is given by the formula:
# Equilibrium = Birthrate / Deathrate - Replicaterate

# THERE IS NO EQUILIBRIUM IF R > D, N --> infinity
birth_chance = 1
death_chance = 0.1
replicate_chance = 0.001
iterations = 10000
cycle_blobs(birth_chance, death_chance, replicate_chance, iterations)
```

Living things break the equilibrium with replication and thus follow the theory of evolution

```python
# THERE IS NO EQUILIBRIUM IF R > D, N --> infinity
# Delta = B + (R - D) * N
birth_chance = 0.01
death_chance = 0.2
replicate_chance = 0.3
iterations = 200
cycle_blobs(birth_chance, death_chance, replicate_chance, iterations)
```
